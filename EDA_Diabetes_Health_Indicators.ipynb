```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("diabetes_012_health_indicators_BRFSS2015.csv")
df.head()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df['BMI'], bins=30, kde=True)
plt.title("Distribution of Body Mass Index (BMI)")
plt.xlabel("BMI")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x=df['BMI'])
plt.title("Boxplot of BMI Distribution")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x=df['PhysHlth'], y=df['MentHlth'])
plt.title("Physical Health vs Mental Health Days")
plt.xlabel("Physical Health (days)")
plt.ylabel("Mental Health (days)")
plt.show()
```

```python
sns.countplot(x=df['Diabetes_012'])
plt.title("Distribution of Diabetes Status")
plt.show()
```
